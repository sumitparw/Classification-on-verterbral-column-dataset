```python
# Rename label names with corresponding numericals
df['lbl'] = df['lbl'].map({'AB': 1, 'NO': 0})
df
```

b)
iii selecting first 70 of NO and first 140 of AB

```python
# Create the dataframe for Training with  70 rows of class0(NO) and 140 rows of class1(AB)
#class 0
df_trainNO = df[df['lbl'] == 0][:70]
# print(df_trainNO)
#class 1
df_trainAB = df[df['lbl'] == 1][:140]

#concatination of both for Training Dataset
df_training = pd.concat([df_trainAB , df_trainNO]).reset_index(drop=True)
df_training
```

```python
# Create the dataframe for Testing with rest rows of class0(NO) and class1(AB)

#class 0
df_testNO = df[df['lbl'] == 0]
df_testNO = df_testNO[70:]

#class 1
df_testAB = df[df['lbl'] == 1]
df_testAB = df_testAB[140:]

#concatination of both for Testing Dataset
df_testing = pd.concat([df_testAB , df_testNO]).reset_index(drop=True)
df_testing
```

```python
# Split Training and Testing dataframes into train_value,test_value,test_label and train_label arrays
train_value = df_training.drop('lbl', axis=1)
train_label = df_training['lbl']



test_value = df_testing.drop('lbl', axis=1)
test_label = df_testing['lbl']

test_value = np.asarray(test_value)
test_label = np.asarray(test_label)

print("Size of:")
print("Training Data", train_value.shape)
print("Testing Data", test_value.shape)
print("Training Label", train_label.shape)
print("Testing Label", test_label.shape)
```

c)
i) Using Sklearn package to implement KNN
ii) k for k âˆˆ {208,205, . . . ,7,4,1,}(in reverse order and find various things

```python
list_k = list()

error_Minimum = float('inf')
final_k = None

errors_Training = list()
errors_Testing = list()

#k âˆˆ {208,205, . . . ,7,4,1,}(in reverse order).

for k in range(208,0,-3):

    neigh = KNeighborsClassifier(n_neighbors=k)
    neigh.fit(train_value, train_label)
    
    list_k.append(k)
    
    error_Training = 1-metrics.accuracy_score(neigh.predict(train_value), train_label)
    error_Testing = 1-metrics.accuracy_score(neigh.predict(test_value), test_label)
    
    errors_Training.append(error_Training)
    errors_Testing.append(error_Testing)
    
    if error_Minimum > error_Testing:
        error_Minimum = error_Testing
        final_k = k
    
print("The Best value for K should be:",final_k)

#The best value of K
```

```python
# plotting of the error for training and testing as in the Graphical view

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(list_k, errors_Training, label="Train Error", color='tab:green')
ax.plot(list_k, errors_Testing, label="Test Error", color='tab:red')
plt.legend(loc='upper right')
plt.grid(True)
plt.xlim(210, 0)
plt.show()
```

```python
#Confusion Matrix when k=k*

neigh = KNeighborsClassifier(n_neighbors=final_k)
neigh.fit(train_value, train_label)

pred_label = neigh.predict(test_value)

confusion_Matrix = metrics.confusion_matrix(test_label, pred_label)
print("Confusion Matrix: \n", confusion_Matrix)
```

```python
#true positive rate when k=kâˆ—.
tn, fp, fn, tp = confusion_Matrix.ravel()
print("True positive rate: ", tp/(tp+fn))
```

```python
#true negative rate when k=kâˆ—.
print("True negative rate: ", tn/(tn+fp))
```

```python
#precision when k=kâˆ—.
print("Precision: ", tp/(tp+fp))
```

```python
#F-score when k=kâˆ—.
print("F1 Score : ", metrics.f1_score(test_label, pred_label))
```

```python
size = list()
errors_testing = list()

#loop starts from 10 goes to 210 with a increment of 10
for i in range(10,220,10):
    
    size.append(i)
    df_train = pd.concat([df_trainNO[:i//3], df_trainAB[:i-i//3]]).reset_index(drop=True)
    
    train_value = df_train.drop('lbl', axis=1)
    train_label = df_train['lbl']

    train_value = np.asarray(train_value)
    train_label = np.asarray(train_label)
    
    best_valofk = None
    error_Minimum = float('inf')   #Initializing with Infinte Number
    

    k = 1
    while k<i:

        neigh = KNeighborsClassifier(n_neighbors=k)
        neigh.fit(train_value, train_label)

        error_testing = 1-metrics.accuracy_score(neigh.predict(test_value), test_label)

        if error_Minimum > error_testing:
            error_Minimum = error_testing
        k += 5
        
    errors_testing.append(error_Minimum)
    

# plotting
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(size, errors_testing, label="Test Error", color='tab:orange')
plt.legend(loc='upper right')
plt.grid(True)
plt.title("Learning Curve")
plt.show()
```

(e) The majority polling decision can be replaced by weighted decision, in which the weight of each point in voting is inversely proportional to its distance from the query/test data point. In this case, closer neighbors of a query point will have a greater influence than neighbors which are further away. Use weighted voting with Euclidean, Manhattan, and Chebyshev distances and report the best test errors when k âˆˆ {1, 6, 11, 16, . . . , 196}.

```python
dist_Measures = ["euclidean", "manhattan", "chebyshev"]

errors_testing = list()
best_Valueofk = list()

for dist_Measure in dist_Measures:

    best_valofk = None
    error_Minimum = float('inf')   #Initializing with Infinte Number

    for k in range(1,210,5):     
        neigh = KNeighborsClassifier(n_neighbors=k, metric=dist_Measure, weights="distance")
        neigh.fit(train_value, train_label)

        error_testing = 1-metrics.accuracy_score(neigh.predict(test_value), test_label)

        if error_Minimum > error_testing:
            error_Minimum = error_testing
            best_valofk = k        
            
    errors_testing.append(error_Minimum)
    best_Valueofk.append(best_valofk)
   
data_frame = {"Distance Measure Type" : dist_Measures, "Best Value of K" : best_Valueofk, "Error In Testing" : errors_testing}
for i in range(0,3,1):
     print ("For distance mesaure", data_frame['Distance Measure Type'][i] ,", Value of K is ", data_frame['Best Value of K'][i] ," and error in testing", round(data_frame['Error In Testing'][i],2))
```
